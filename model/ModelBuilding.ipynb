```python
def add_company(inpt):
    if inpt == 'Moto' or inpt == 'Realme' or inpt == 'Motorola' or inpt == 'Nothing'or inpt == 'Nexus'or inpt == 'A10E'or inpt == 'a'or inpt == '�9A'or inpt == '10A':
        return 'Other'
    else:
        return inpt
```

```python
x=data.drop('Price_In_LKR',axis=1)
y=data['Price_In_LKR']
```

```python
data.head(2)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

Model Trainning

```python
def model_acc(model):
    model.fit(x_train, y_train)
    acc = model.score(x_test, y_test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

Hyper parameter tunning

```python
from sklearn.model_selection import GridSearchCV



parameters = {'criterion': ['friedman_mse', 'poisson'],
              'max_depth': [ 10, 20, 30]}


grid_obj = GridSearchCV(estimator=dt, param_grid=parameters, scoring='r2', cv=5, error_score='raise')


grid_fit = grid_obj.fit(x_train, y_train)

best_model = grid_fit.best_estimator_
test_score = best_model.score(x_test, y_test)

print("Best Hyperparameters:", grid_fit.best_params_)
print("Test Score:", test_score)

```

Model Saving

```python
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
